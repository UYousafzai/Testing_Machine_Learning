```python
import pandas as pd 
from bs4 import BeautifulSoup
import re
import nltk
import numpy as np
```

```python
train = pd.read_csv('abalone.csv', header=0, delimiter=',', quoting=3) 
display(train)
```

```python
def tokenize1(row_review):
    row_review = re.sub('[M]',"0.1",row_review)
    row_review = re.sub('[F]',"0.2",row_review)
    row_review = re.sub('[I]',"0.3",row_review)
    return row_review
   

train_copy = train.copy()
train_copy.ix[:,0] = train_copy.ix[:,0].apply(tokenize1)

display(train_copy)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_copy.ix[:,0:8], train_copy.ix[:,8], test_size=0.20)
display(train_copy.ix[:,0:8])
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear')
clf.fit(X_train, y_train) 
```

```python
clf.decision_function(X_test)
predictions = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score =accuracy_score(np.array(y_test), predictions)
print score
```
